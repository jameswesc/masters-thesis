```python
import pandas as pd
import numpy as np
import geopandas as gpd
import xarray as xr

import pyvista as pv
pv.set_jupyter_backend('client')

from pathlib import Path
import ipywidgets as widgets
from IPython.display import display
```

```python
plot_lidar_dir = Path("../data/outputs/plots/lidar")
plots_gdf = gpd.read_file("../data/outputs/plots/plots.geojson")
# plots_gdf
plots = plots_gdf["id"].to_list()

# Create interactive widgets for site selection
plot_dropdown = widgets.Dropdown(
    options=plots,
    value=plots[0] if plots else None,
    description='plot:',
    disabled=False,
)

metrics_dir = Path("../data/outputs/plots/metrics")

def read_metrics(plot_id: str, xy_bin_size: float | None = None, z_bin_size : float | None = 1):
    dir = metrics_dir / f"xy_{xy_bin_size}_z_{z_bin_size}"
    path = dir / f"{plot_id}.nc"

    return xr.load_dataset(path)
```

```python
plot_dropdown = widgets.Dropdown(
    options=plots,
    value=plots[0] if plots else None,
    description='plot:',
    disabled=False,
)
display(plot_dropdown)
```

# Datasets

```python
metrics = read_metrics(plot_dropdown.value, xy_bin_size=5)
metrics
```

# Height

```python
def show_cover(plot):
    metrics = read_metrics(plot, xy_bin_size=1, z_bin_size=1)
    return metrics['max_veg'].T.plot()

out = widgets.interactive_output(show_cover, {'plot': plot_dropdown})
display(out)
```

# Cover

```python
def show_cover(plot):
    metrics = read_metrics(plot, xy_bin_size=1, z_bin_size=1)
    # return metrics['lcover'].T.plot()
    return metrics['pct_inside_z'].sel(z=slice(15,None)).sum(dim='z').T.plot()

out = widgets.interactive_output(show_cover, {'plot': plot_dropdown})
display(out)
```
