```python
import pandas as pd
import numpy as np
```

```python
site_info = pd.read_csv("../data/source/environmental_data__site_information.csv")
site_info = site_info.rename(columns={
    "Master_ID": "site",
    "Class": "site_type",
    "Disturbance": "disturbance",
    "Estabyear": "year"
})
site_info = site_info[['site', 'site_type', 'disturbance', 'year']]
site_info['year'] = site_info['year'].fillna(-9999).astype('int')

# Add site_number column: sequential numbering of unique site_ids within each site_type
site_info['site_number'] = site_info.groupby('site_type')['site'].transform(
    lambda x: pd.factorize(x)[0] + 1
)
site_info = site_info.set_index('site')
site_info
```

```python
plot_metrics = pd.read_parquet("../data/outputs/plots/plot_metrics.parquet")
plot_metrics['plot_number'] = plot_metrics.index.str[-1].astype(int)
plot_metrics
```

```python
plot_metrics_with_site_info = plot_metrics.join(site_info.drop(columns='site_type'), on='site')
plot_metrics_with_site_info
```

```python
plot_metrics_with_site_info.to_parquet("../data/outputs/plots/plot_metrics_with_site_info.parquet")
```
