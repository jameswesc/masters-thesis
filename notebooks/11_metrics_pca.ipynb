```python
from pathlib import Path

import pandas as pd

outputs_dir = Path('../data/outputs')
```

```python
plot_sumaries = pd.read_csv(outputs_dir / "plot_summary_metrics.csv")
plot_sumaries = plot_sumaries.set_index('id')
plot_sumaries = plot_sumaries.drop(columns=['mean__point_density', 'mean__pulse_density', 'mean__scan_angle_mean'])
plot_sumaries
```

```python
# Import libraries for PCA analysis
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer

# Set up plotting style
plt.style.use('default')
sns.set_palette("husl")
```

```python
# Scale and center the metrics
scaler = StandardScaler()
scaled_metrics = scaler.fit_transform(plot_sumaries)

# First fit PCA with full set of components
# and find out how many components needed to explain 95% of the variance
pca = PCA()
pca.fit(scaled_metrics)
cumulative_variance = np.cumsum(pca.explained_variance_ratio_)
n_components_95 = np.argmax(cumulative_variance >= 0.95) + 1

pca = PCA(n_components=n_components_95)
pca.fit(scaled_metrics)
pca_scores = pca.transform(scaled_metrics)


pca_scores_df = pd.DataFrame(
    pca_scores,
    columns=pca.get_feature_names_out(),
    index=plot_sumaries.index
)

pca_scores_df
```

```python
pca_loadings_df = pd.DataFrame(pca.components_.T, columns=pca.get_feature_names_out(), index=plot_sumaries.columns)
pca_loadings_df.index.name = 'metric'
pca_loadings_df
```

```python
pca_scores_df.to_csv(outputs_dir / "pca_scores.csv")
pca_loadings_df.to_csv(outputs_dir / "pca_loadings.csv")
```
